```python
import pandas as pd
from Bio import SeqIO
import os
from collections import defaultdict
```

### P26-4C

```python
sequences = SeqIO.index("C:\\Users\\eliss\\Downloads\\P26-4C-FrogSkin_S11_L001_R1_001.fastq", "fastq")
sizes = defaultdict(int)
for _id in sequences.keys():
    length = len(sequences[_id].seq)
    sizes[length] += 1

df = pd.Series(data=sizes)
df.index.name = "Size"
df = df.to_frame().reset_index().rename(columns= {0: 'Count'})
```

```python
df.sort_values(by="Count", ascending=False)
```
